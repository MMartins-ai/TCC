```python
names = ['PL21_22.csv', 'PL20_21.csv', 'PL19_20.csv', 'PL18_19.csv','PL17_18.csv', 'PL16_17.csv', 'PL15_16.csv', 'PL14_15.csv', 'PL13_14.csv', 'PL12_13.csv', 'PL11_12.csv', 'PL10_11.csv','PL09_10.csv']
```

```python
def matriz_inc_points(data, name):
    f= lambda x: x[:5].upper()
    try:
        data['HomeTeam'].replace('Man City', 'City', inplace=True)
        data['AwayTeam'].replace('Man City', 'City', inplace=True)
    except:
        print("City it's not here")
    try:
        data['HomeTeam'].replace('West Brom', 'Brom', inplace=True)
        data['AwayTeam'].replace('West Brom', 'Brom', inplace=True)
    except:
        print("West Brom it's not here")
    try:
        data['HomeTeam'].replace('Blackpool', 'Pool', inplace=True)
        data['AwayTeam'].replace('Blackpool', 'Pool', inplace=True)
    except:
        print("Blackpool it's not here")
    data['HomeTeam'] = data['HomeTeam'].map(f)
    data['AwayTeam'] = data['AwayTeam'].map(f)
    data['HomePoints'] = np.nan
    data.loc[data['FTHG']>data['FTAG'], 'HomePoints'] = 3
    data.loc[data['FTHG']==data['FTAG'], 'HomePoints'] = 1
    data.loc[data['FTHG']<data['FTAG'], 'HomePoints'] = 0
    matriz1 = data.pivot('AwayTeam','HomeTeam','HomePoints')
    matriz1.fillna(0, inplace=True)
    data['AwayPoints'] = np.nan
    data.loc[data['FTHG']>data['FTAG'], 'AwayPoints'] = 0
    data.loc[data['FTHG']==data['FTAG'], 'AwayPoints'] = 1
    data.loc[data['FTHG']<data['FTAG'], 'AwayPoints'] = 3
    matriz2 = data.pivot('HomeTeam','AwayTeam','AwayPoints')
    matriz2.fillna(0, inplace=True)
    matriz = matriz1+matriz2
    matriz.to_csv(name)
```

```python
for i,j in zip(lista,names):
    matriz_inc_points(i,j)
```
