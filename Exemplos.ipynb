```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
G.add_weighted_edges_from([(1,2,3),(4,5,6),(7,8,2),(1,7,1),(2,4,3),(8,5,4),(8,1,3),(8,4,10)])
```

```python
all_weights = []
for (node1,node2,data) in G.edges(data=True):
    all_weights.append(data['weight'])
```

```python
unique_weights = list(set(all_weights))
```

```python
pos = nx.spring_layout(G)
```

```python
for weight in unique_weights:
    weighted_edges = [(node1,node2) for (node1,node2, edge_atrr) in G.edges(data=True) if edge_atrr['weight']==weight]
    width = weight
    nx.draw_networkx_edges(G,pos ,edgelist = weighted_edges, width=width)
```
